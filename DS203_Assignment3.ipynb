```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dsdata3.csv')
df.head()
```

```python
x=df.x
y=df.y
Xo=np.array(x)
Yo=np.array(y)
```

```python
plt.scatter(Yo,Xo)
plt.show()
```

```python
XY=np.multiply(Xo,Yo)
XY_=np.mean(XY)
x_=np.mean(Xo)
y_=np.mean(Yo)
x_y_=x_*y_
x2=np.multiply(Xo,Xo)
x2_=np.mean(x2)
a=(XY_-x_y_)/(x2_-x_**2)
b=(y_-a*x_)
print(a,b)
```

```python
X=np.vstack((Xo,Yo)).T
```

```python
y_predicted=a*Xo+b
plt.scatter(Yo,Xo)
plt.plot(y_predicted,Xo,color='red')
plt.show()
```

```python
y_error=y_predicted-Yo
y_error
```

```python
from scipy import stats
res=stats.normaltest(y_error)
p=res.pvalue
p
```

```python
SST=np.sum((Yo-y_)**2)
SSR=np.sum((y_predicted-y_)**2)
SSE=np.sum((y_predicted-Yo)**2)
SST
```

```python
SSE+SSR
```
